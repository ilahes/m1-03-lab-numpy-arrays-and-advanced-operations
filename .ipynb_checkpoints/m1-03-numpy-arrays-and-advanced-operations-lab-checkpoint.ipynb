```python
#task 1
import numpy as np
np.random.seed(42)
readings = np.random.normal(50, 5, (360, 4))

print(f"Shape: {readings.shape} | Dtype: {readings.dtype}")
print("First 3 columns:\n", readings[:3])
```

### task 1 
In this task, we create a 2D NumPy array `readings` representing six hours of minute-level readings from 4 sensors (shape `(360, 4)`). We use `np.random.normal` to simulate normally distributed sensor values around 50 with a standard deviation of 5.  
Setting a fixed seed (`np.random.seed(42)`) ensures reproducibility of results.  
The first three rows and the shape/dtype of the array are printed to verify that the data structure is correct.

```python
#task 2
import numpy as np

scaled_readings = (readings - readings.min()) / (readings.max() - readings.min())
centered_readings = readings - readings.mean(axis=0)

print(f"Are all shapes same? {readings.shape == scaled_readings.shape == centered_readings.shape}")
print(f"Scaled Min/Max: {scaled_readings.min():.1f}, {scaled_readings.max():.1f}")
```

### task 2
We apply two common vectorized transformations on the sensor data:
Scaling to 0–1 (`scaled_readings`): Using the formula `(X - min) / (max - min)`, all values are rescaled based on the global minimum and maximum of the dataset. This is fully vectorized — no loops are used.  

Centering (`centered_readings`): Each column (sensor) is centered by subtracting its mean. Using `axis=0` ensures the subtraction is column-wise, keeping the array shape unchanged.  

Finally, we verify that both transformed arrays have the same shape as the original `readings`.

```python
#3
import numpy as np
outlier_mask = (readings > 80) | (readings < 20)

outlier_count = np.sum(outlier_mask)
outlier_percentage = (outlier_count / readings.size) * 100

cleaned_readings = readings.copy()
cleaned_readings[outlier_mask] = np.nan
nan_count = np.isnan(cleaned_readings).sum()

print(f"Number of outliers: {outlier_count}")
print(f"Outlier percentage: {outlier_percentage:.2f}%")
print(f"Is the number of NaNs equal to the number of outliers? {nan_count == outlier_count}")
```

### task 3
We identify outliers in the sensor data as values greater than 80 or less than 20.  

- A boolean mask `outlier_mask` is created using elementwise comparisons and the `|` operator for logical OR.  
- The total number of outliers and their percentage relative to all readings are calculated.  
- Outliers are replaced with `np.nan` in a copy of the original data (`cleaned_readings`) to preserve the original matrix.  
- We verify that the number of NaNs matches the computed outlier count.  
This approach avoids loops and leverages vectorized boolean indexing.

```python
 #4
import numpy as np
sensor_means = readings.mean(axis=0)
sensor_stds = readings.std(axis=0)
time_means = readings.mean(axis=1)

print(f"Sensor means shape: {sensor_means.shape} (Gözlənilən: 4)")
print(f"Sensor standard deviations shape: {sensor_stds.shape} (Gözlənilən: 4)")
print(f"Time means shape: {time_means.shape} (Gözlənilən: 360)")

print("\nMean of the first 2 sensors:", sensor_means[:2])
```

### task 4
We compute summary statistics:
Sensor-wise (column-wise) summaries: `sensor_means` and `sensor_stds` are computed with `axis=0` to summarize each sensor across all time steps.  
Time-wise (row-wise) summaries: `time_means` is computed with `axis=1` to get the average reading across sensors at each time step.  

Validation checks confirm that `sensor_means` and `sensor_stds` have length 4 (one per sensor) and that `time_means` has length 360 (one per time step).


```python
#5 
report = {
    "total_readings": readings.size,
    "outlier_count": outlier_count,
    "outlier_percent": outlier_percentage,
    "sensor_means": sensor_means,
    "sensor_stds": sensor_stds
}

print("--- Part 1: Validation Report ---")
for key, value in report.items():
    print(f"{key}: {value}")

# Sadə bir yoxlama (Validation)
if report["total_readings"] == readings.size:
    print("\n Total readings count is correct")
```

### task 5
We create a dictionary `report` that collects key metrics from Part 1:
- `total_readings`: total number of readings (`readings.size`)  
- `outlier_count` and `outlier_percent`  
- `sensor_means` and `sensor_stds`  

The dictionary is printed in a readable format. A simple validation check confirms that `total_readings` matches the size of the original `readings` array.

```python
# Task 6: Create a reproducible simulation matrix
rng = np.random.default_rng(seed=42)
sim = rng.normal(0, 1, (1000, 6))

print("--- Task 6: Simulation Matrix ---")
print(f"Shape: {sim.shape} | Dtype: {sim.dtype}")
print("First 3 rows:\n", sim[:3])
```

### task 6
We generate a simulation matrix `sim` representing 1000 trials for 6 scenarios.  
- `np.random.default_rng(seed=42)` initializes a reproducible random number generator.  
- `rng.normal(0, 1, (1000, 6))` produces a standard normal matrix.  
The shape, dtype, and first three rows are printed for verification.


```python
# Task 7: Apply broadcasting for scenario adjustments
# 1D array of offsets for 6 scenarios
scenario_shift = np.array([-0.3, -0.1, 0.0, 0.1, 0.2, 0.4])

# Broadcasting: (1000, 6) + (6,) -> Adding shift to each column
adjusted_sim = sim + scenario_shift

print("\n--- Task 7: Broadcasting Check ---")
print(f"Adjusted Sim shape: {adjusted_sim.shape}")
print(f"Original column means: {sim.mean(axis=0).round(2)}")
print(f"Adjusted column means: {adjusted_sim.mean(axis=0).round(2)}")
```

### task 7
We apply small scenario-specific offsets using broadcasting:
- `scenario_shift` is a 1D array of length 6 containing offsets for each scenario.  
- Adding this array to `sim` via `adjusted_sim = sim + scenario_shift` leverages NumPy broadcasting to apply each offset column-wise.  

The resulting `adjusted_sim` keeps the same shape as `sim`. The column means shift in the direction of the offsets, demonstrating the effect of broadcasting.


```python
# Task 8: Rank scenarios with sorting and partitioning

scenario_means = adjusted_sim.mean(axis=0)
sorted_indices = np.argsort(scenario_means)
partitioned_values = np.partition(scenario_means, -2)[-2:]
full_sort_top_two = np.sort(scenario_means)[-2:]
partition_check = np.allclose(np.sort(partitioned_values), full_sort_top_two)

print("--- Task 8: Ranking and Partitioning ---")
print(f"Scenario Means: {scenario_means.round(4)}")
print(f"Sorted Indices (Lowest to Highest): {sorted_indices}")
print(f"Top 2 Values (from partition): {partitioned_values.round(4)}")
print(f"Does partition match full sort? {partition_check}")
```

### task 8 
We analyze and rank the scenarios:
1. Compute mean outcome per scenario: `scenario_means = adjusted_sim.mean(axis=0)`  
2. Rank scenarios from lowest to highest using `sorted_indices = np.argsort(scenario_means)`  
3. Identify the top two values using `np.partition(scenario_means, -2)[-2:]`  
4. Verify that these top two values match the two largest values from a full sort (`np.sort(scenario_means)[-2:]`).  

This task demonstrates efficient ranking and partial sorting without fully sorting the array.

```python
# Task 9: Controlled randomness and reproducibility

rng_2 = np.random.default_rng(seed=42)
sim_2 = rng_2.normal(0, 1, (1000, 6))

rng_3 = np.random.default_rng(seed=99)
sim_3 = rng_3.normal(0, 1, (1000, 6))

# Verification checks
matches_same_seed = np.array_equal(sim, sim_2)
differs_diff_seed = not np.array_equal(sim, sim_3)

print("--- Task 9: Reproducibility Check ---")
print(f"sim_2 matches sim (same seed): {matches_same_seed}")
print(f"sim_3 differs from sim (different seed): {differs_diff_seed}")
```

### task 9
We test reproducibility with seeds:
- Re-generate `sim_2` with the same seed (42) → should exactly match `sim`  
- Generate `sim_3` with a different seed (99) → should differ from `sim`  

Verification is done using `np.array_equal` and simple logical checks.

```python
# Task 10: Build a validation report (Part 2)

# Identify the top two indices and values for the report
# Using the sorted_indices from Task 8
top_two_indices = sorted_indices[-2:].tolist()
top_two_values = scenario_means[top_two_indices].tolist()

# Create a report dictionary
report_part_2 = {
    "shape": adjusted_sim.shape,
    "scenario_means": scenario_means.round(4).tolist(),
    "top_two_indices": top_two_indices,
    "top_two_values": [round(v, 4) for v in top_two_values]
}

# Convert report to a readable string and print
print("--- Part 2: Simulation Validation Report ---")
for key, value in report_part_2.items():
    print(f"{key}: {value}")

# Final validation statements
valid_shape = adjusted_sim.shape == (1000, 6)
valid_means_len = len(scenario_means) == 6
valid_top_indices_len = len(top_two_indices) == 2

print("\n--- Final Checks ---")
print(f"Is shape (1000, 6)? {valid_shape}")
print(f"Does scenario_means have length 6? {valid_means_len}")
print(f"Does top_two_indices have length 2? {valid_top_indices_len}")
```

### task 10
We create a validation report for Part 2 containing:
- `shape` of the adjusted simulation matrix  
- `scenario_means` rounded for readability  
- `top_two_indices` and `top_two_values` based on previous ranking  

We print the report in a readable format. Final validation checks confirm that the shape, scenario means length, and number of top indices are correct.  
This ensures the simulation and ranking workflow is reproducible and consistent.
